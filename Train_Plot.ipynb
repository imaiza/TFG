## Representacion para ver rango de nuestro algoritmo 

```python
def bineador2(dfp, targetname,target, nbins):
    
    #target el nuevo M500
    #dfp2=dfp.sort_values(by=target)
    dfp2=dfp
    
    X=(dfp2['M500c(41)']).values
    Y=(dfp2[targetname]/X).values
    Z=(dfp[target]/X).values
    Y=(Y-Z)*100/Z
    
    bmean, bedge, bnumber= stats.binned_statistic(X, Y, statistic='mean',bins=nbins) #bar values
    yerr, bedge2, bnumber2= stats.binned_statistic(X, Y, statistic='std',bins=nbins) #standard deviation values
    
    bincenters=0.5*(bedge[1:]+bedge[:-1])
    return bincenters, bmean, yerr

def bineador22(dfp, targetname,target, nbins):
    
    #target el nuevo M500
    #dfp2=dfp.sort_values(by=target)
    dfp2=dfp
    
    X=(dfp2['M500c(41)']).values
    Y=(dfp2[targetname]).values
    Z=(dfp[target]).values
    Y=(Y-Z)*100/Z
    
    bmean, bedge, bnumber= stats.binned_statistic(X, Y, statistic='mean',bins=nbins) #bar values
    yerr, bedge2, bnumber2= stats.binned_statistic(X, Y, statistic='std',bins=nbins) #standard deviation values
    
    bincenters=0.5*(bedge[1:]+bedge[:-1])
    return bincenters, bmean, yerr
```
